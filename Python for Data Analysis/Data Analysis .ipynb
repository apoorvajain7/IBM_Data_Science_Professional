```python
#BIN "GROUPING THE PRICES IN BIN"

import numpy as np
bins = np.linspace(min(df["price"]),max(df["price"]),4)
group_name=["low","medium","high"]
df["price_binning"]=pd.cut(df["price"], bins, labels= group_name)
```

```python
#LINEAR REGRESSION MODELLING
#IMPORTING SKLEARN
from sklearn.linear_model import LinearRegression
```

```python
#CREATE OBJECT USING CONSTRUCTOR
lm = LinearRegression()
```

```python
x=df[['highway-mpg']]
y=df['price']
```

```python
#FITTING THE MODEL
lm.fit(x,y)

#PREDICT THE X
yhat=lm.predict(x)

#b0,b1 
b0=lm.intercept_
b1=lm.coef_

print('the intercept is ', b0)
print('the slope is', b1)
```

```python
df = df.reset_index()
```

```python
#MULTIPLE LINEAR REGRESSION

df[["price"]] = df[["price"]].astype("float")
z=df[['horsepower','engine-size','curb-weight','highway-mpg']]
#fitting

X=lm.fit(z, df['price'])

#predict
lm.predict(X)
```

```python
#RESIDUAL PLOT ITS AN ERROR BETWEEN ACTUAL VALUE AND PREDICTED VALUE

sns.residplot(df['highway-mpg'], df['price'])

#the result shows a curvature therfore the relationship is not linear 
```

```python
sns.residplot(df['wheel-base'], df['price'])
#the reult is linear because no curvature here 
```

```python
#DISTRIBUTION PLOT

ax1=sns.distplot(df['price'],hist=False, color='r', label="Actual Value")
sns.distplot(yhat,ax=ax1, hist=False, color='b', label="Fitted Value")
```
